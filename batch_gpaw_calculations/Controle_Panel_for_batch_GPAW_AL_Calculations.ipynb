```python

import ase
from ase.io.trajectory import Trajectory
from ase.visualize import view

from ase.md.velocitydistribution import MaxwellBoltzmannDistribution
from ase.md.verlet import VelocityVerlet
from ase.md.langevin import Langevin
from ase import units

from ase.calculators.calculator import Calculator, all_changes
import nequip
from nequip.ase import NequIPCalculator

import os
import time

import numpy as np
import pandas as pd

import torch

import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import rc
%matplotlib inline


import seaborn as sns

from matplotlib import rc
import matplotlib.ticker
rc={'font.size' : 11, 
    'font.family' : 'computer modern',
    'axes.labelsize': 11,   
    'legend.fontsize': 11, 
    'axes.titlesize': 11, 
    'xtick.labelsize': 11, 
    'ytick.labelsize': 11}
sns.set(font = 'computer modern', rc=rc)
sns.set_style("ticks")

# **************************************************************************************
# Name of input trajectory file:
input_traj_name = 'combined.traj'
# Run name:
run_name = 'AL_configs_1'
# **************************************************************************************

# Getting number of input atoms:
len_traj = len( Trajectory('Input_trajectory/' + input_traj_name) )
# For writing in the py file:

# For writing in files for submition:
input_traj_string = 'Trajectory(\'Input_trajectory/combined.traj\')['

for i in range(len_traj):
    
    with open('Templet_files/GPAW_PBC_x_y_not_z.py', 'r') as file:
        
        GPAW_file = file.read()
        
        GPAW_file = GPAW_file.replace('AL_JOB_NAME',
                    'AL_' + str(i))
        
        GPAW_file = GPAW_file.replace('--input_trajectory--',
                            input_traj_string + str(i) + ']')
        
        GPAW_file = GPAW_file.replace('--output_trajectory--',
                    'Trajectory(\'GPAW_output/GPAW_out_' + str(i) + '\', \'w\')')
        
        GPAW_file_name = run_name + '_' + str(i) + '.py'
        with open(GPAW_file_name, 'w') as file:
            file.write(GPAW_file)
        
        print('! gpaw sbatch ' + GPAW_file_name)
        
print('! squeue -u s173973')
```

```python
# Getting all configurations in fodler: AL_1_configs:


import ase
from ase.io.trajectory import Trajectory
from ase.visualize import view

from ase.md.velocitydistribution import MaxwellBoltzmannDistribution
from ase.md.verlet import VelocityVerlet
from ase.md.langevin import Langevin
from ase import units

from ase.calculators.calculator import Calculator, all_changes
import nequip
from nequip.ase import NequIPCalculator

import os
import time

import numpy as np
import pandas as pd

import torch

import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import rc
%matplotlib inline



traj_file_names = os.listdir("GPAW_output")[1:]



AL_atoms = []
for traj_file_name in traj_file_names:
    
    try:
        traj_file = Trajectory("GPAW_output/" + traj_file_name)
    
        for atoms in traj_file:
        
            AL_atoms.append(atoms)
            
    except:
        pass
    

print(len(AL_atoms))
```
